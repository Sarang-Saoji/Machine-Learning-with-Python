```python
#Deleting Unnnecessary Columns
#Dropping the column "dish_liked", "phone", "url" and saving the new dataset as "zomato"
```

```python
zomato1 = zomato.drop(['url','phone','dish_liked'],axis=1)
```

```python
zomato1.columns
```

```python
#Removing the Duplicates
#looking at the dataset after transformation
```

```python
zomato1 = zomato1.drop_duplicates()
```

```python
zomato1.head()
```

```python
#Remove the NaN values from the dataset
#.info() function is used to get a concise summary of the dataframe
```

```python
zomato1.isnull().sum()
zomato1 = zomato1.dropna(how = 'any')
```

```python
zomato1.info()
```

```python
zomato1.columns
```

```python
#Changing the column names
zomato1 = zomato1.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
```

```python
zomato1.columns
```

```python
#Some Transformations
zomato1['cost'] = zomato1['cost'].astype(str)
zomato1['cost'] = zomato1['cost'].apply(lambda x : x.replace(',','.') )
zomato1['cost'] = zomato1['cost'].astype(float)
zomato1.info()
```

```python
#Reading uninque values from the Rate column
```

```python
zomato1['rate'].unique()
```

```python
#Removing '/5' from Rates
```

```python
zomato1 = zomato1.loc[zomato1.rate != 'NEW']
zomato1 = zomato1.loc[zomato1.rate != '-'].reset_index(drop=True)
```

```python
zomato1['rate'].unique()
```

```python
zomato1['rate'] = zomato1['rate'].apply(lambda x : x.replace('/5','')).str.rstrip()
zomato1['rate'] = zomato1['rate'].astype(float)
```

```python
zomato1.info()
zomato1.head()
```

```python
zomato1['name']
```

```python
# Adjust the column names
```

```python
zomato1['online_order'] = zomato1['online_order'].replace(('Yes','No'),(True,False))
```

```python
zomato1.head()
```

```python
zomato1['book_table'] = zomato1['book_table'].replace(('Yes','No'),(True, False))
```

```python
zomato1.head()
```

```python
zomato1['name'] = zomato1['name'].apply(lambda x : x.title())
```

```python
zomato1.head()
```

```python
#Encode the input Variables
 # looking at the dataset after transformation
```

```python
def Encode(zomato):
    for column in zomato.columns[~zomato.columns.isin(['rate', 'cost', 'votes'])]:
        zomato[column] = zomato[column].factorize()[0]
    return zomato

zomato_en = Encode(zomato1.copy())
zomato_en.head()
```

```python
import seaborn as sns
```

```python
#Get Correlation between different variables
corr = zomato_en.corr(method='kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
zomato_en.columns
```
